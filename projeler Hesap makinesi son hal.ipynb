```python
from tkinter import*
master=Tk()
canvas=Canvas(master,height=400,width=250)
canvas.pack()
frame_ust=Frame(master,bg='black')
frame_ust.place(relx=0, rely=0, relheight=0.3,relwidth=1)
frame_alt=Frame(master,bg='black')
frame_alt.place(relx=0, rely=0.301, relheight=0.7,relwidth=1)

sonuc=Entry(frame_ust, width=18,  bg = 'black',justify=RIGHT, fg = '#c4c1c1', font=('Medieval 17 bold'), )
sonuc.place(x=10,y=90)


def press(x):
    s=len(sonuc.get())
    return sonuc.insert(s,str(x))
hesaplar=' '
def esit():
    try:
        global hesaplar
        hesaplar=sonuc.get()
        total = str(eval(hesaplar))
        sonuc.delete(0,'end')
        sonuc.insert(0,total)
        return 
    except:
        sonuc.delete(0,'end')
        hata="Geçersiz işlem"
        s=len(sonuc.get())
        sonuc.insert(s,str(hata))

def temizle():
    sonuc.delete(0,'end')
    return
def geri():
    s=len(sonuc.get())
    sonuc.delete(s-1,str(s))
    return 
parantez_sayısı=1
def parantez():
    global parantez_sayısı
    s=len(sonuc.get())
    if s==0:
        sonuc.insert(s,str('('))
        parantez_sayısı=2
    elif (s!=0) and (parantez_sayısı==1):
        sonuc.insert(s,str('*('))
        parantez_sayısı=2
        
    elif (s!=0) and (parantez_sayısı==2):
        parantez_sayısı=1
        sonuc.insert(s,str(')'))
            
            
sıfır_butonu=Button(frame_alt,text='0',activebackground = "#4E4848",command =lambda: press(0),relief=FLAT, bg='#4E4848',fg='white',font='Medieval 12 bold')
#sıfır_butonu.pack(padx=1,pady=180,side=BOTTOM)
sıfır_butonu.place(relx=0.2345, rely=0.9, relheight=0.1,relwidth=0.22)

bir_butonu=Button(frame_alt,text='1',activebackground = "#4E4848",command =  lambda: press(1),relief=FLAT, bg='#4E4848',fg='white',font='Medieval 12 bold')
bir_butonu.place(relx=0.01, rely=0.69, relheight=0.2,relwidth=0.22)

iki_butonu=Button(frame_alt,text='2',activebackground = "#4E4848",command =  lambda: press(2),relief=FLAT, bg='#4E4848',fg='white',font='Medieval 12 bold')
iki_butonu.place(relx=0.2345, rely=0.69, relheight=0.2,relwidth=0.22)


uc_butonu=Button(frame_alt,text='3',activebackground = "#4E4848",command =   lambda: press(3),relief=FLAT, bg ='#4E4848',fg='white',font='Medieval 12 bold')
uc_butonu.place(relx=0.46, rely=0.69, relheight=0.2,relwidth=0.22)



dort_butonu=Button(frame_alt,text='4',activebackground = "#4E4848",command = lambda: press(4),relief=FLAT, bg='#4E4848',fg='white',font='Medieval 12 bold')
dort_butonu.place(relx=0.01, rely=0.47, relheight=0.2,relwidth=0.22)


bes_butonu=Button(frame_alt,text='5',activebackground = "#4E4848",command =  lambda: press(5),relief=FLAT, bg='#4E4848',fg='white',font='Medieval 12 bold')
bes_butonu.place(relx=0.2345, rely=0.47, relheight=0.2,relwidth=0.22)


    
altı_butonu=Button(frame_alt,text='6',activebackground = "#4E4848",command = lambda: press(6),relief=FLAT,  bg='#4E4848',fg='white',font='Medieval 12 bold')
altı_butonu.place(relx=0.46, rely=0.47, relheight=0.2,relwidth=0.22)


yedi_butonu=Button(frame_alt,text='7',activebackground = "#4E4848",command = lambda: press(7),relief=FLAT,  bg='#4E4848',fg='white',font='Medieval 12 bold')
yedi_butonu.place(relx=0.01, rely=0.25, relheight=0.2,relwidth=0.22)


sekiz_butonu=Button(frame_alt,text='8',activebackground = "#4E4848",command =lambda: press(8), relief=FLAT, bg='#4E4848',fg='white',font='Medieval 12 bold')
sekiz_butonu.place(relx=0.2345, rely=0.25, relheight=0.2,relwidth=0.22)

    
dokuz_butonu=Button(frame_alt,text='9',activebackground = "#4E4848",command =lambda: press(9) ,relief=FLAT, bg='#4E4848',fg='white',font='Medieval 12 bold')
dokuz_butonu.place(relx=0.46, rely=0.25, relheight=0.2,relwidth=0.22)
###İŞLEMLER
artı_butonu=Button(frame_alt,text='+',activebackground = "#4E4848",command = lambda: press('+'),relief=FLAT, bg='#4E4848',fg='#ff7b00',font='Medieval 12 bold')
artı_butonu.place(relx=0.69, rely=0.73, relheight=0.15,relwidth=0.3)

eksi_butonu=Button(frame_alt,text='-',activebackground="#4E4848",command=lambda:press('-') ,relief=FLAT,bg="#4E4848",fg='#ff7b00',font='Medieval 12 bold')
eksi_butonu.place(relx=0.69, rely=0.56, relheight=0.15,relwidth=0.3)

carpma_butonu=Button(frame_alt,text='*',activebackground="#4E4848",command=lambda:press('*') ,relief=FLAT,bg="#4E4848",fg='#ff7b00',font='Medieval 12 bold')
carpma_butonu.place(relx=0.69, rely=0.39, relheight=0.15,relwidth=0.3)

bolme_butonu=Button(frame_alt,text='/',activebackground="#4E4848",command=lambda:press('/') ,relief=FLAT,bg="#4E4848",fg='#ff7b00',font='Medieval 12 bold')
bolme_butonu.place(relx=0.69, rely=0.22, relheight=0.15,relwidth=0.3)

nokta_butonu=Button(frame_alt,text='.',activebackground="#4E4848",command=lambda:press('.') ,relief=FLAT,bg="#4E4848",fg='white',font='Medieval 12 bold')
nokta_butonu.place(relx=0.46, rely=0.9, relheight=0.1,relwidth=0.22)


esittir_butonu=Button(frame_alt,text='=',activebackground = "#4E4848",command = esit,relief=FLAT, bg='#ed9d09',fg='white',font='Medieval 12 bold')
esittir_butonu.place(relx=0.69, rely=0.05, relheight=0.16,relwidth=0.3)

temizle_butonu=Button(frame_alt,text='C',activebackground = "#4E4848",command = temizle, relief=FLAT, bg='#4E4848',fg='red',font='Medieval 12 bold')
temizle_butonu.place(relx=0.01, rely=0.05, relheight=0.18,relwidth=0.336)

geri_butonu=Button(frame_alt,text='<',activebackground = "#4E4848",command = geri,relief=FLAT, bg='#4E4848',fg='red',font='Medieval 12 bold')
geri_butonu.place(relx=0.35, rely=0.05, relheight=0.18,relwidth=0.336)

parantez_butonu=Button(frame_alt,text='()',activebackground="#4E4848",command=parantez, relief=FLAT,bg="#4E4848",fg='white',font='Medieval 12 bold')
parantez_butonu.place(relx=0.01, rely=0.9, relheight=0.1,relwidth=0.22)


master.mainloop()
```
